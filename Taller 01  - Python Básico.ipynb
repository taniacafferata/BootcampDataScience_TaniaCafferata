## Estructuras de Control (Condicionales y Ciclos)
Haz un programa que ayude a automatizar procesos repetitivos y aplicar reglas de negocio (lógica condicional).

1. Crea un ciclo for que recorra tu lista de registro_ventas (puedes agregar un par de ventas manuales extra al código para hacer la prueba).
2. Dentro del ciclo, aplica un condicional if: Si el total de la venta es mayor a $100000, aplica un descuento del 10%.
3. Imprime el reporte final mostrando el producto, el total original y el total con descuento (si aplica).

```python
## Se crean unas ventas extras

registro_venta.extend([
    {"producto": "juguete", "categoria": "Mascotas", "precio": 10000, "cantidad":10, "total": 100000},
    {"producto": "Licuadora", "categoria": "Electronica", "precio": 80000, "cantidad":1, "total": 80000}
])


## ciclo for que recorre la lista registro_venta

for producto in registro_venta:
    total= int(producto["precio"]) * int(producto["cantidad"])

    #Se crea un condicional para indicar si la venta es mayor a $100000 y asi aplicar un descuento del 10%

    if total > 100000:
        descuento = total * 0.10
        total_final = total - descuento #Se hace formula para indicar el nuevo valor total
        #Se imprime mensaje indicando el descuento
        print(f"producto: {producto['producto']} - total original: ${total} - nuevo total: $ {total_final} (se aplico 10% de descuento)")

    else:
        total_final = total
        #se imprime mensaje cuando no se realiza descuento
        print(f"producto: {producto['producto']} total: $ {total_final} (sin descuento)")

 
```
